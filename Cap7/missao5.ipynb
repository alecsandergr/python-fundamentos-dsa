```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Carrega o arquivo
load_file = '/content/drive/MyDrive/Python Fundamentos DSA/Cap 07/Dados/dados_compras.json'
purchase_file = pd.read_json(load_file, orient = "records")
purchase_file.head()
```

## Informações Sobre os Consumidores

### Número total de consumidores

Como cada compra de item é registrada, deve-se obter os valores únicos da variável `Login`

```python
dados_consumidores = purchase_file.loc[:, ['Login', 'Idade', 'Sexo']].drop_duplicates()
dados_consumidores
```

```python
total_consumidores = pd.DataFrame({'Total de Consumidores': [dados_consumidores.count()[0]]})
total_consumidores
```

## Análise Geral de Compras

* Número de itens exclusivos
* Preço médio de compra
* Número total de compras
* Rendimento total

```python
itens_exclusivos = purchase_file['Item ID'].unique()
itens_exclusivos
```

```python
print(f'O número de itens exclusivos é: {itens_exclusivos.shape[0]}')
```

```python
resumo = pd.DataFrame()
resumo['Preço Médio'] = [round(purchase_file['Valor'].mean(),2)]
resumo['Soma das Vendas'] = [purchase_file['Valor'].sum()]
resumo['Contagem das Vendas'] = [purchase_file['Valor'].count()]
resumo['Itens únicos'] = [itens_exclusivos.shape[0]]
resumo.head()
```

## Informações Demográficas Por Gênero

* Porcentagem e contagem de compradores masculinos
* Porcentagem e contagem de compradores do sexo feminino
* Porcentagem e contagem de outros / não divulgados

```python
contagem_genero = dados_consumidores['Sexo'].value_counts()
percentagem_genero = (contagem_genero/total_consumidores.iloc[0,0])*100
resumo_demografia = pd.DataFrame({'Sexo': contagem_genero,
                                  'Percentagem': round(percentagem_genero,2)})
resumo_demografia
```

## Análise de Compras Por Gênero

* Número de compras
* Preço médio de compra
* Valor Total de Compra
* Compras for faixa etária

```python
resumo_genero = purchase_file.groupby('Sexo')['Valor'].agg(['sum', 'count', 'mean']).sort_values('count', ascending=False)
resumo_genero.rename(columns = {'sum': 'Valor Total de Compra',
                                'count': 'Número de compras', 
                                'mean': 'Preço Médio'}, inplace=True)
resumo_genero['Preço Médio'] = round(resumo_genero['Preço Médio'],2)

# O total normalizao conta apenas os clientes únicos
resumo_genero['Total Normalizado'] = round(resumo_genero['Valor Total de Compra']/resumo_demografia['Sexo'],2)
resumo_genero
```

Compras por faixa etária

```python
valores_fe = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 150]
texto_fe = ['Menos de 10', '10 a 14', '15 a 19', '20 a 24', '25 a 29', '30 a 34', '35 a 39', 'Mais de 40']
purchase_file['Faixa Etária'] = pd.cut(purchase_file['Idade'], bins = valores_fe, labels=texto_fe)
purchase_file
```

```python
resumo_fe = purchase_file.groupby('Faixa Etária')['Valor'].agg(['sum', 'mean', 'count']).sort_values('count', ascending=False)
resumo_fe.rename(columns = {'sum': 'Valor Total de Compra',
                            'count': 'Número de compras', 
                            'mean': 'Preço Médio'}, inplace=True)

resumo_fe['Preço Médio'] = round(resumo_fe['Preço Médio'],2)
resumo_fe['Percentagem de Compras'] = round((resumo_fe['Número de compras']/total_consumidores.iloc[0,0])*100,2)


resumo_fe = resumo_fe.sort_index()
resumo_fe
```

## Consumidores Mais Populares (Top 5)

 Pelo valor total de compra

* Login
* Número de compras
* Preço médio de compra
* Valor Total de Compra

```python
top_5_clientes = purchase_file.groupby('Login')['Valor'].agg(['count', 'mean', 'sum'])
top_5_clientes = top_5_clientes.sort_values('sum', ascending=False)
top_5_clientes.rename(columns = {'count': 'Número de Compras', 
                                 'mean': 'Preço Médio', 
                                 'sum': 'Valor Total'}, inplace=True)
top_5_clientes.head()
```
